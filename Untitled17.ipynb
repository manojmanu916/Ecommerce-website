```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.feature_selection import RFE, SelectKBest, f_classif
import matplotlib.pyplot as plt
import seaborn as sns
import joblib


"""
Required Datasets (Available on UCI Machine Learning Repository):
1. Heart Disease Dataset:
   - URL: https://archive.ics.uci.edu/ml/datasets/heart+disease
   - File: processed.cleveland.data (for heart disease prediction)

2. Diabetes Dataset:
   - URL: https://archive.ics.uci.edu/ml/datasets/Early+stage+diabetes+risk+prediction+dataset
   - File: diabetes_data_upload.csv

3. Breast Cancer Dataset:
   - URL: https://archive.ics.uci.edu/ml/machine-learning-databases/00529/diabetes_data_upload.csv - O diabetes.csv

4. Breast Cancer Dataset:
   - URL: https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data - O breast_cancer.csv

Download these files and place them in a 'data' folder in your project directory.
"""

def load_data(dataset_name):
    if dataset_name == 'heart':
        columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
                  'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
        data = pd.read_csv('/content/heart.csv', header=None, names=columns, na_values='?')
        data['target'] = data['target'].apply(lambda x: 1 if x > 0 else 0)
        target_column = 'target'

    elif dataset_name == 'diabetes':
        data = pd.read_csv('/content/diabetes.csv')
        for col in data.columns:
            if data[col].dtype == 'object':
                data[col] = data[col].astype('category').cat.codes
        target_column = data.columns[-1]

    elif dataset_name == 'breast_cancer':
        columns = ['id', 'diagnosis'] + [f'feature_{i}' for i in range(1, 31)]
        data = pd.read_csv('/content/breast_cancer.csv', header=None, names=columns)
        data['diagnosis'] = data['diagnosis'].map({'M':1, 'B':0})
        target_column = 'diagnosis'

    else:
        raise ValueError("Invalid dataset name. Choose from 'heart', 'diabetes', or 'breast_cancer'")

    return data, target_column

def preprocess_data(data, target_column):
    data = data.dropna() if data.isnull().sum().sum() > 0 else data
    X = data.drop(target_column, axis=1)
    y = data[target_column]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    return X_scaled, y, scaler

def feature_selection(X, y, method='rfe', k=10):
    if method == 'rfe':
        estimator = LogisticRegression(max_iter=1000)
        selector = RFE(estimator, n_features_to_select=k)
        selector.fit(X, y)
        selected_features = selector.support_
    elif method == 'kbest':
        selector = SelectKBest(f_classif, k=k)
        selector.fit(X, y)
        selected_features = selector.get_support()
    else:
        selected_features = np.ones(X.shape[1], dtype=bool)

    return X[:, selected_features], selected_features

def train_evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1': f1_score(y_test, y_pred),
        'cv_mean': cross_val_score(model, X_train, y_train, cv=5).mean()
    }

    plt.figure(figsize=(6,4))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
    plt.title(f'Confusion Matrix - {type(model).__name__}')
    plt.show()

    return metrics, model

def hyperparameter_tuning(X, y, model, param_grid):
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grid,
        cv=5,
        scoring='accuracy',
        n_jobs=-1
    )
    grid_search.fit(X, y)
    return grid_search.best_params_, grid_search.best_score_

def predict_new_sample(model, scaler, sample, selected_features=None):
    if selected_features is not None:
        sample = sample[selected_features]
    sample_scaled = scaler.transform(sample.reshape(1, -1))
    prediction = model.predict(sample_scaled)
    probability = None
    if hasattr(model, 'predict_proba'):
        probability = model.predict_proba(sample_scaled)[0]
    return prediction[0], probability

def main():
    print("\nAvailable datasets:")
    print("1. Heart Disease (cleveland)")
    print("2. Diabetes")
    print("3. Breast Cancer")

    while True:
        choice = input("Select dataset (1-3): ").strip()
        if choice in ['1', '2', '3']:
            break
        print("Invalid choice. Please try again.")

    datasets = {'1': 'heart', '2': 'diabetes', '3': 'breast_cancer'}
    dataset_name = datasets[choice]

    data, target_column = load_data(dataset_name)
    X, y, _ = preprocess_data(data, target_column)

    X_selected, selected_features = feature_selection(X, y, k=10)
    print(f"\nSelected {sum(selected_features)} important features")

    X_train, X_test, y_train, y_test = train_test_split(
        X_selected, y, test_size=0.3, random_state=42)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    models = {
        'Logistic Regression': LogisticRegression(max_iter=1000),
        'SVM': SVC(probability=True),
        'Random Forest': RandomForestClassifier(),
        'XGBoost': XGBClassifier()
    }

    results = {}
    print("\nTraining models...")
    for name, model in models.items():
        print(f"\nEvaluating {name}...")
        metrics, trained_model = train_evaluate_model(model, X_train_scaled, X_test_scaled, y_train, y_test)
        results[name] = metrics
        models[name] = trained_model

        if hasattr(model, 'feature_importances_'):
            plt.figure(figsize=(10,4))
            plt.bar(range(len(model.feature_importances_)), model.feature_importances_)
            plt.title(f'Feature Importance - {name}')
            plt.show()

    print("\nModel Evaluation Results:")
    results_df = pd.DataFrame(results).T
    print(results_df)

    best_model_name = results_df['accuracy'].idxmax()
    print(f"\nPerforming hyperparameter tuning for {best_model_name}...")

    param_grids = {
        'Logistic Regression': {'C': [0.001, 0.01, 0.1, 1, 10], 'penalty': ['l2'], 'solver': ['lbfgs']},
        'SVM': {'C': [0.1, 1, 10], 'gamma': [0.1, 0.01], 'kernel': ['rbf']},
        'Random Forest': {'n_estimators': [100, 200], 'max_depth': [None, 10, 20]},
        'XGBoost': {'n_estimators': [100, 200], 'max_depth': [3, 6], 'learning_rate': [0.01, 0.1]}
    }

    best_params, best_score = hyperparameter_tuning(
        X_train_scaled, y_train,
        models[best_model_name],
        param_grids[best_model_name]
    )

    print(f"Best parameters: {best_params}")
    print(f"Best validation score: {best_score:.4f}")

    final_model = models[best_model_name].__class__(**best_params)
    final_model.fit(X_train_scaled, y_train)

    sample_idx = np.random.randint(0, X_test_scaled.shape[0])
    sample = X_test_scaled[sample_idx]
    pred, proba = predict_new_sample(final_model, scaler, sample)

    print("\nExample Prediction:")
    print(f"Sample index: {sample_idx}")
    print(f"Predicted class: {'Positive' if pred == 1 else 'Negative'}")
    if proba is not None:
        print(f"Class probabilities: {np.round(proba, 3)}")
    else:
        print("Class probabilities not available for this model.")

    model_filename = f'best_model_{dataset_name}.pkl'
    joblib.dump(final_model, model_filename)
    print(f"\nModel saved as {model_filename}")

if __name__ == "__main__":
    main()
```
